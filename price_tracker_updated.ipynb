```python
#SCRAPPING DATA FROM AMAZON 
```

```python
url1="https://www.amazon.in/s?k=iphone+12&crid=3POH9FEZM4GZR&sprefix=iphone+%2Caps%2C215&ref=nb_sb_ss_pltr-ranker-retrain-acsession-acceptance_1_7"
resultx=requests.get(url1)
soupx=BeautifulSoup(resultx.text,"lxml")
content2x=soupx.find_all('div',class_="a-section a-spacing-small a-spacing-top-small")
for item in content2x:
    content3x=item.find_all('span',class_=["a-size-base-plus a-color-base a-text-normal","a-size-medium a-color-base a-text-normal"])
    content4x=item.find_all('span',class_="a-price-whole")
    content5x=item.find_all('a',class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal")
    if len(content3x)==0:
        lx="Not Available Model"
    else:
        lx=str(content3x[0]).split(">")[1][:-6]
       
   
    if len(content4x)==0:
       px="Not Available Price"
    else:
        px=str(content4x[0]).split(">")[1][:-6]
    l_ama_model.append(lx)
    l_ama_price.append(px)
    if len(content3x)==0:
        lix="Not Available Model"
    else:
        lix=str(content5x[0].get('href'))
    l_ama_link.append("https://www.amazon.in"+lix)   

```

```python
for i in range(x_result):
    if result_xxx.iloc[:,-3][i] !="Not Available Price" and result_xxx.iloc[:,-3][i]!="NO UPDATE" and result_xxx.iloc[:,-5][i]!="Not Available Price" :
        result_xxx["PRICE_UPDATE"][i]=int(result_xxx.iloc[:,-3][i])-int(result_xxx.iloc[:,-5][i])
```

```python
#Highlighting the rows with yellow color in which price decrement is more than 100 rupees
```

```python
def highlight_greaterthan(x):
    if x.PRICE_UPDATE<-100:
        return ['background-color: yellow']*6
    else:
        return ['background-color: white']*6
result_xxx=result_xxx.style.apply(highlight_greaterthan,axis=1)        
```
